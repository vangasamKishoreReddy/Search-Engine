```python
chroma_db['doc_vector_pretrained_bert'] = chroma_db['doc_vector_pretrained_bert'].str.replace('[', '').str.replace(']', '')
```

```python
chroma_db
```

```python
import pandas as pd
from tqdm import tqdm

# Assuming sub_title and df are defined elsewhere

for i, row in tqdm(chroma_db.iterrows(), total=len(chroma_db), desc="Processing rows"):
    embedding = row['doc_vector_pretrained_bert']
    if pd.notnull(embedding):  # Check if embedding is not null
        ids = f"chunk_{i}"  # Unique ID for each chunk
        
        # Split the string representation of the embedding and convert each part to float
        embedding_list = [float(x) for x in embedding.split()]
        
        sub_title.add(ids,  # Add to collection
                       embeddings=embedding_list,  # Use the converted list
                       documents=row["file_content_chunks"],  # Use chunk text as document
                       metadatas={"name": row["name"]}  # Use 'name' as metadata
                       )
    else:
        # Fill null values with zeros
        row['doc_vector_pretrained_bert'] = [0.0] * len(row['doc_vector_pretrained_bert'])
```
